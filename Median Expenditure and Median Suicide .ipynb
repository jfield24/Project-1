```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import os
import json
import time
from sklearn import preprocessing
import gmaps
import gmaps.datasets
import gmaps.geojson_geometries

from matplotlib.cm import viridis
from matplotlib.colors import to_hex

# Import API key
from api_key import g_key
```

```python
descriptives_table = pd.read_csv("./output_data/descriptives_table.csv")
country_coordinates = pd.read_csv("./country_co-ordinates.csv")

combined_df = descriptives_table.merge(country_coordinates, on="Country", how="left")

# descriptives_table.count()[0
combined_df.sort_values("Median_Health_Expenditure")
```

```python
#Configuring gmaps
gmaps.configure(api_key=g_key)

#Store latitude and longitude in locations
locations = combined_df[["latitude", "longitude"]]

#Store humidity in humidity
expenditure = combined_df["Median_Health_Expenditure"]
```

```python
# Plotting Heatmap
fig = gmaps.figure(center=(46.0, -5.0), zoom_level=2)
max_intensity = np.max(expenditure)

#Creating heat layer
heat_layer = gmaps.heatmap_layer(locations, weights=expenditure, dissipating=False,
                                     max_intensity=max_intensity, opacity=0.8, point_radius=10)

#Adding layer
fig.add_layer(heat_layer)

#Displaying figure
fig
```

```python
#Number of Suicide 
sucide_rate = combined_df["Median_Suicide_Rate"]

# Plotting Heatmap
fig = gmaps.figure(center=(46.0, -5.0), zoom_level=2)
max_intensity = np.max(sucide_rate)

#Creating heat layer
heat_layer = gmaps.heatmap_layer(locations, weights=sucide_rate, dissipating=False,
                                     max_intensity=max_intensity, opacity=0.6, point_radius=10)

#Adding layer
fig.add_layer(heat_layer)

#Displaying figure
fig
```
