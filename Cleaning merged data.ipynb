```python
# Import dependencies
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
# Read in the CSV file with the merged dataset
merged_df = pd.read_csv("data_sources/merged_data.csv", encoding='utf-8')
merged_df
```

```python
# Find '..' values - not nulls but also useless - remove them
dots = merged_df.loc[merged_df["health_expenditure"] == ".."].index
merged_df.drop(dots, inplace=True)

# Drop null values from the dataset
merged_df = merged_df.dropna(how='any')

```

```python
# Change the suicide column to be suicide rate per 100,000 people
merged_df["suicide_rate (%)"] = 1000 * merged_df["suicide_rate (%)"]

# Rename the suicide_rate (%) column
merged_df.rename(columns = {'suicide_rate (%)':'number of suicides (per 100,000 people)', 'health_expenditure': 'health expenditure (per capita)'}, inplace = True)
```

```python
# Write the merged clean data into a csv file
merged_df.to_csv("output_data/dataset_complete.csv", index=False)
```
